```python
# Standardizing the Amount column(All other columns are already scaled)
from sklearn.preprocessing import RobustScaler

rob_scale= RobustScaler()  # Robust to  outliers
df['Scaled_Amount']=rob_scale.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
# Dropping column'Amount' from dataset
df.drop(['Amount'],axis=1,inplace=True)

scaled_amount=df['Scaled_Amount']
df.drop('Scaled_Amount',axis=1,inplace=True)
df.insert(0,'Scaled_Amount',scaled_amount)
df.head()
```

```python
# Let Shuffle the data before creating the subsamples
# Fraud vs. legit(genuine) transactions
Fraud_Transactions=df[(df['Class']==1)]
legit_Transactions=df[(df['Class']==0)][:492]
```

```python
# Sampling dataset and concatenating fraud and legit transaction
df=df.sample(frac=1)
normal_distributed_df=pd.concat([Fraud_Transactions,legit_Transactions])
dfs=normal_distributed_df.sample(frac=1,random_state=42)

dfs.head()
```

```python
# Distribution of class in subsample dataset
print('Distribution of Classes in the subsample dataset')
print(dfs['Class'].value_counts()/len(dfs))
```

```python
# Checking correlation between features on the balanced dataset
f,ax2=plt.subplots(figsize=(24,10))

corr=dfs.corr()
sns.heatmap(corr,cmap='coolwarm_r',annot_kws={'size':20},ax=ax2)

ax2.set_title("Balanced Correlation Matrix", fontsize=14)
```

# SPLITTING THE DATASET

```python
# Splitting data into X and Y
X=dfs.drop(columns='Class')
y=dfs['Class']
```

```python
# X dataframe
X.shape
```

```python
# Y dataframe 
y.shape
```

```python
# Applying Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
# X_train dimension
X_train.shape
```

```python
# X_test dimension
X_test.shape
```

# APPLYING MODELS

```python
# Importing different models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
models={"Logistic Regression":LogisticRegression(),
        "SVC":SVC(),
        "Decision Tree":DecisionTreeClassifier()}
```

```python
# Cross validation score of models
from sklearn.model_selection import cross_val_score

for name,classifier  in models.items():
    classifier.fit(X_train,y_train)
    training_score=cross_val_score(classifier,X_train,y_train,cv=5)
    print(f"Classifier:{name} has a training score of {np.mean(training_score)*100:.2f}% accuracy")
```

# HYPERTUNING PARAMETERS WITH GRID SEARCH CV

```python
# Hypertuning of parameters of models using GridSearch CV
from sklearn.model_selection import GridSearchCV

# Logistic Regression parameters hypertuning
log_paras ={"penalty":['l1','l2'],'C':[0.001,0.01,0.1,1,10,100,1000]}
grid_search_log =GridSearchCV(LogisticRegression(),log_paras)
grid_search_log.fit(X_train,y_train)
log= grid_search_log.best_estimator_

# Support Vector Classifier parameters hypertuning
svc_paras ={'kernel':['rbf','poly','sigmod','linear'],'C':[0.5,0.7,0.9,1,10]}
grid_search_svc =GridSearchCV(SVC(),svc_paras)
grid_search_svc.fit(X_train,y_train)
svc= grid_search_svc.best_estimator_

# Decision tree Classifier parameters hypertuning
tree_paras ={"criterion":["gini","entropy"],"max_depth":list(range(2,4,1)),"min_samples_leaf":list(range(5,7,1))}
grid_search_tree =GridSearchCV(DecisionTreeClassifier(),tree_paras)
grid_search_tree.fit(X_train,y_train)
tree= grid_search_tree.best_estimator_
```

```python
# Hypertuning Results
print(log,'\n',svc,'\n',tree)
```

# CROSS - VALIDATION OF MODELS

```python
# cross_val_score after hypertuning of models

# Logistic Regression
log_score=cross_val_score(log,X_train,y_train,cv=5)
print('Logistic Regression has cross validation score:',round(log_score.mean()*100,2).astype(str)+'%')

# Support Vector Classifier
svc_score=cross_val_score(svc,X_train,y_train,cv=5)
print('SVC cross has validation score:',round(svc_score.mean()*100,2).astype(str)+'%')

# Decision Tree Classifier
tree_score=cross_val_score(tree,X_train,y_train,cv=5)
print('Decision Tree has cross validation score:',round(tree_score.mean()*100,2).astype(str)+'%')
```

```python
SVC and Logistic Regression has best cross validation score
```

```python
# Cross Validation Predictions

from sklearn.model_selection import cross_val_predict

# Logistic Rgression Prediction
log_pred= log.predict(X_test)

# Support Vector Classifier Prediction
svc_pred= svc.predict(X_test)

# Decision Tree Classifier Prediction
tree_pred=tree.predict(X_test)
```

```python
# ROC-AUC Score of models
from sklearn.metrics import roc_auc_score

print('Logistic Regression:', roc_auc_score(y_test,log_pred))

print('SVC:', roc_auc_score(y_test,svc_pred))

print('Decision Tree:', roc_auc_score(y_test,tree_pred))
```

```python
# predictions of test data
y_pred_log= log.predict(X_test)
y_pred_svc=svc.predict(X_test)
y_pred_tree=tree.predict(X_test)
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix

# Confusion matrix of Logistic Regression
log_cm=confusion_matrix(y_test,y_pred_log)


# Confusion matrix of Support Vector Classifier
svc_cm=confusion_matrix(y_test,y_pred_svc)

# Confusion matrix of Decision Tree Classifier
tree_cm=confusion_matrix(y_test,y_pred_tree)
```

```python
# Heatmap of Confusion Matrix

fig,ax=plt.subplots(2,2,figsize=(22,12))

# Plotting Logistic Regression confusion matrix 
sns.heatmap(log_cm,ax=ax[0][0],annot=True)
ax[0,0].set_title("Logistic Regression \n Confusion Matrix",fontsize=14)
ax[0,0].set_xticklabels(['',''],fontsize=14,rotation=90)
ax[0,0].set_yticklabels(['',''],fontsize=14,rotation=90)


#  Plotting Support Vector Classifier confusion matrix 
sns.heatmap(svc_cm,ax=ax[0][1],annot=True)
ax[0,1].set_title("Support Vector Machine \n Confusion Matrix",fontsize=14)
ax[0,1].set_xticklabels(['',''],fontsize=14,rotation=90)
ax[0,1].set_yticklabels(['',''],fontsize=14,rotation=90)

#  Plotting Decision Tree Classifier confusion matrix 
sns.heatmap(tree_cm,ax=ax[1][0],annot=True)
ax[1,0].set_title("Decision Tree \n Confusion Matrix",fontsize=14)
ax[1,0].set_xticklabels(['',''],fontsize=14,rotation=90)
ax[1,0].set_yticklabels(['',''],fontsize=14,rotation=90)

fig.delaxes(ax[1][1])
plt.tight_layout()
plt.show()
```

```python
# Classification Report of Models
from sklearn.metrics import classification_report

print('Logistic Regression')
print(classification_report(y_test,y_pred_log))

print('Support Vector Classifier')
print(classification_report(y_test,y_pred_svc))

print('Decision Tree Classifier')
print(classification_report(y_test,y_pred_tree))
```
